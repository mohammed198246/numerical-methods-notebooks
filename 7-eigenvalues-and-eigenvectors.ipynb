```python
using LinearAlgebra
```

```python
# Power method
function power_rule(A, x, iterations)
    y = x
    for n = 1:iterations
        y = A * y
        y = y / norm(y, Inf)
    end
    y 
end
```

```python
A = randn(5,5)
x = randn(5)
```

```python
power_rule(A, x, 100)
```

```python
A = [2 1 ; 1 2]
x = [1 ; 0]
```

```python
power_rule(A, x, 10)
```

```python
A * power_rule(A, x, 10)
```
